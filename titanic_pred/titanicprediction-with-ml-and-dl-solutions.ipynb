# Preprocessing for DATA

```python
def preprocessing_data(data):
    
    #overall information about the data
    print(data.describe())
    print('NaNs Values in the data:\n',data.isna().sum())
    
    #Dropping Cabin from data
    #**For Cabin: It is supposed to have 891 values but 687 of them is NaN. So I made the decition for it to remove from the data so that it does not effect to my model.****
    data = data.drop('Cabin', axis=1) 
    
    # For Age: fill NaNs with mean value
    mean_age = int(data['Age'].mean())
    data['Age']  = [mean_age if pd.isna(each) else each for each in data['Age']]
    
    # For Age: fill NaNs with mean value
    mean_age = int(data['Fare'].mean())
    data['Fare']  = [mean_age if pd.isna(each) else each for each in data['Fare']]
    
    #for Embarked: filling NaNs with the most embarked place
    value_counts = data['Embarked'].value_counts()
    # Get the key (index) with the maximum value
    max_key = value_counts.idxmax()
    #filling NaNs with max_keys
    data['Embarked'] = [max_key if pd.isna(each) else each for each in data['Embarked']]
    
    print('After process: NaNs Values in the data:\n',data.isna().sum())
    
    #For Prefix:
    prefix = []
    prefix = [each.split(' ')[1] for each in data['Name']]
    name_list = data['Name'][0] 
    name_list.split(' ')[1]
    
    data['Prefix'] = prefix
    
    ###  Note: I want to put them into different category such as 
    ###  
    
    male = ['Mr.', 'Master.', 'Major.', 'Col.', 'Capt.']
    female = [ 'Mrs.', 'Miss.', 'Mme.', 'Ms.', 'Mlle.']
    professional = [ 'Dr.', 'Rev.']
    nobility = ['Jonkheer.', 'Don.']
    miscellaneous = [ 'y', 'the', 'der']

    #put numeric values into each categories:::
    # ---- male(0), female(1), and neutral categories(2), noble titles(3), military titles(4), regional titles(5), 
    # Use a single map operation to assign category values
    data['Prefix'] = data['Prefix'].map(lambda x: 
                                    0 if x in male else 
                                    1 if x in female else 
                                    2 if x in professional else 
                                    3 if x in nobility else 
                                    4 if x in miscellaneous else 
                                    5)
    
    data['Embarked'].replace('S', 1, inplace=True)
    data['Embarked'].replace('C', 2, inplace=True)
    data['Embarked'].replace('Q', 3, inplace=True)
    
    data['Sex'].replace('male', 0, inplace=True)
    data['Sex'].replace('female', 1, inplace=True)
    
    return data
```

# Train the model with Machine Learning Solution (RandomForestClassifier)

```python
from sklearn.preprocessing import StandardScaler
# Standardize the data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
import joblib



def model_evaluater(X_train,y_train,model_name,mymodel):
    
    # Initialize and train the Random Forest model
    #rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
    mymodel.fit(X_train, y_train)

    # Make predictions on the validation set
    y_val_pred_rf = mymodel.predict(X_test)
    accuracy = accuracy_score(y_test, y_val_pred_rf)
    print("Accuracy for {0}:{1}".format(model_name,accuracy))
    
    # Save the model to a file
    joblib.dump(mymodel,'/kaggle/working/random_forest_model.pkl')

    
    return y_val_pred_rf
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import ConfusionMatrixDisplay

def conv_matrix(y_test, y_val_pred_rf,name):
    # Compute the confusion matrix
    cm = confusion_matrix(y_test, y_val_pred_rf)

    # Plot the confusion matrix using ConfusionMatrixDisplay
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Not-Living','Alive'])
    disp.plot(cmap='Blues')
    plt.title('Confusion Matrix for {}'.format(name))
    plt.show()
    
```

```python
first_model = {
    'RandomForestClassifier': RandomForestClassifier(n_estimators=100, random_state=42)
}

for k,v in first_model.items():
    y_val_pred_rf = model_evaluater(X_train,y_train,k,v)
    conv_matrix(y_test, y_val_pred_rf,k)
    
```

# What if FL is applied to model

```python
from tensorflow.keras import Input
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.callbacks import ModelCheckpoint
from keras.optimizers import Adam
```

```python
feature
```

```python
model = Sequential()
from tensorflow.keras.layers import Dropout


# Use Input layer to define the input shape
model.add(Input(shape=(len(feature),)))
# Add hidden and output layers
model.add(Dense(64,activation = 'softplus'))  # Second hidden layer
model.add(Dense(32,activation = 'relu'))  # Second hidden layer
model.add(Dropout(0.5))
model.add(Dense(16,activation = 'softplus'))  # Second hidden layer
model.add(Dense(1,activation ='sigmoid')) # 1 unit for binary classification
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate = 0.001), metrics=['accuracy'])
```

```python
#saving the best weight for best accuracy. so i need to use CheckPoint:
from tensorflow.keras.callbacks import ModelCheckpoint

filepath = 'titanic.weights.h5'

model_checkpoint_callback = ModelCheckpoint(
    filepath = filepath,
    save_weights_only=True,
    monitor='accuracy',
    mode='max',
    verbose = 1,
    save_best_only=True)

callbacks_list = [model_checkpoint_callback]

history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test),callbacks = callbacks_list, verbose = 1)
```

```python
# Accessing the history dictionary
history_dict = history.history

# Print all the available metrics
print(history_dict.keys())

print('accuracy:',max(history_dict['accuracy']))
print('loss:',min(history_dict['loss']))
print('val_accuracy:',max(history_dict['val_accuracy']))
print('val_loss:',min(history_dict['val_loss']))
```
