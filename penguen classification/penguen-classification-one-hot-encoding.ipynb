```python
#this def helps out changing NaN to Mean Value of column that are being worked on.

def nan_to_mean(mylist):
    non_nan = list(filter(lambda x: not np.isnan(x),mylist))
    mean_value = round(np.mean(non_nan),3)
    
    filled_list = [mean_value if np.isnan(x) else x for x in mylist]
    return filled_list
# it is returning the list that fitted with mean value instead of NaN
```

```python
#this def helps out that if encounters non_sense value such as '.' or '+' ,'_' etc, it will fill with 'NaN' value so that it will be completed 
#it with mean value so that the weight of the feeture is not disrupted.
def char_to_nan(mylist):
    temp_list= df['Sex']
    my_nan = temp_list[3]
    temp_list = [my_nan if item == '.' else item for item in temp_list]

    for each in range(len(temp_list)):
        if temp_list[each] == '.':
            print(each)
    mylist = temp_list
    return mylist
```

```python
#let see which column/feature has missing values and how much? According to that info, we need missing features so that it goes to nan_to_mean methods.
def missing_value (mylist):
    missing_values = mylist.isna().sum()
    nan_list =[]
    print("Missing values in each column:")
    for k,v in missing_values.items():
        print(f"{k}: {v / df.shape[0] * 100:.2f}%")
        if v != 0:
            nan_list += [k]
    return nan_list
```

```python
#here, what i aimed is to see what kind of outputs/varients are there in categorical features: after that, i need to make them as numerical values
#so i can get them into training process of my model.

non_numerical = ['Species','Region','Island','Stage','Individual ID','Sex']

def varienty_search(feature,mylist):
    mycheck = []
    for each in mylist:
        if each not in mycheck:
            mycheck += [each] 
    print(feature,':',mycheck)


for each in (df[non_numerical]):
    varienty_search(each, df[each])
```

```python
#preprocessing for non_numerical features. on the other words, it makes them as categorical
def preprocessing(df):
    Species_class = {'Adelie Penguin (Pygoscelis adeliae)': 0,
                 'Chinstrap penguin (Pygoscelis antarctica)':1, 
                 'Gentoo penguin (Pygoscelis papua)':2}
    df['Species']= [Species_class[each] for each in df['Species']]

    
    Region_class = {'Anvers':0}
    df['Region']= [Region_class[each] for each in df['Region']]

    Island_class = {'Torgersen':0,
                'Biscoe':1, 
                'Dream':2}
    df['Island']= [Island_class[each] for each in df['Island']]

    Stage_class = {'Adult, 1 Egg Stage':0}
    df['Stage']= [Stage_class[each] for each in df['Stage']]

    Clutch_class = {'Yes':0, 'No':1}
    df['Clutch Completion']= [Clutch_class[each] for each in df['Clutch Completion']]

    counter = 0
    Individual_ID = []
    for each in range(len(df['Individual ID'])):
        Individual_ID += [counter] 
        counter+=1
    df['Individual ID'] = Individual_ID
    
    dp = df['Sex']
    for counter in range(len(dp)):
        each = dp.loc[counter]
        if each == 'FEMALE':
            dp.at[counter] = 0
        elif each == 'MALE':
            dp.at[counter] = 1
    df['Sex'] = dp
```

```python
'''
In machine learning, it is common practice to scale features so that they have a similar range of values. This helps to prevent features with larger ranges from dominating
those with smaller ranges, which can lead to biased or suboptimal models.
'''

from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict, KFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score


y = df.Species
features = ['Region', 'Island', 'Stage', 'Clutch Completion', 
            'Culmen Length (mm)', 'Culmen Depth (mm)', 'Flipper Length (mm)', 
            'Body Mass (g)', 'Sex', 'Delta 15 N (o/oo)', 'Delta 13 C (o/oo)']
X = scaled_df[features]

#Preprocessing pipeline

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), ['Culmen Length (mm)', 'Culmen Depth (mm)', 'Flipper Length (mm)', 'Body Mass (g)', 'Delta 15 N (o/oo)', 'Delta 13 C (o/oo)']),
        ('cat', OneHotEncoder(), ['Region', 'Island', 'Stage', 'Clutch Completion', 'Sex'])
    ])

# Create a pipeline with preprocessing and the classifier
model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(random_state=42))])

# Set up K-Fold Cross-Validation
kfold = KFold(n_splits=5, random_state=42, shuffle=True)

# Perform cross-validation and get predictions
predictions = cross_val_predict(model, X, y, cv=kfold)

# Perform cross-validation and get accuracy scores
scores = cross_val_score(model, X, y, cv=kfold, scoring='accuracy')

# Output the results
print(f"Accuracy for each fold: {scores}")
print(f"Average Accuracy: {scores.mean()}")

# Output predictions
print("Predictions:")
print(predictions)

# Evaluate the predictions
print(f"Overall Accuracy: {accuracy_score(y, predictions)}")
```

```python
'''
lets see the feature importances:
When you use cross-validation, each fold involves fitting a new model on a subset of the data, so the model in the pipeline is not fitted on the entire dataset at once. 
Therefore, you cannot directly access feature_importances_ from the model object because the fitting process occurs separately for each fold. Now, i need to re-do fitting process in each fold
then make each ones average.
'''


#last important thing is that as using k-folds, importance will exist in each fold. so i need to average them.

def get_feature_importances_from_cv(model, X, y, cv):
    from sklearn.base import clone
    importance = []
    
# cv.split(X, y) provides the indices that will be assigned to the training and testing sets for each fold in the K-fold cross-validation process.
    for train_index, test_index in cv.split(X,y):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        # right now, it is splitted form. 
        
        #as this is recurring until the last fold, the model has a tricky issue. if I would not refresh model that already fitted at the last fold, the output would be biased because of the last process.!!!
        #in order to prevent from having biased output, we need to CLONE model so that it ensure us to use model as if it is used in first time.
        # biased output can be generated by overfitting or underfitting which is not desired in ML.
        
        cloned_model = clone(model)
        cloned_model.fit(X_train,y_train)
      
       

        
        #just to refresh my memory
        model = Pipeline(steps=[
            ('preprocessor', preprocessor),
            ('classifier', RandomForestClassifier(random_state=42))])
       
        
        rf_classifier = cloned_model.named_steps['classifier']  # This is how to reach out any elements in pipeline by just writing named_steps ['xxx']
        weights = rf_classifier.feature_importances_
        print(weights)
        importance.append(weights)    # it will be filled with new importances by each k-fold
        
    # Average the feature importances across all folds
    avg_importances = np.mean(importance, axis=0)
    
    return avg_importances
```
